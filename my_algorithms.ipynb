# Factorial

```python
def my_factorial(input):
    factorial=1
    i=1
    while i<=input:
        factorial = factorial * i
        i = i+1
    return factorial
```

```python
print(my_factorial(5))
```

# Linear search

```python
def linear_search(arr, target):
    for index,element in enumerate(arr):
        if element == target:
            return index

    return -1
#It searches for the element to be searched for by its index and compares each element with the searched element. For larger lists, this is very slow. 
#Therefore, this method is less efficient.Maximum search steps equal is all elements number.
```

```python
arr = ["aasdad","basfsfasf","aasdasd"]
target="basfsfasf"
print(linear_search(arr,target))
```

# Binary search

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```

# Array

An array is a fundamental data structure used in programming that allows you to store a collection of elements of the same type. It provides a way to organize and access data in a contiguous block of memory.

In an array, elements are stored in consecutive memory locations, and each element is accessed by its index or position within the array. The index starts at 0 for the first element and increments by one for each subsequent element.

Arrays can be one-dimensional, two-dimensional, or multi-dimensional. A one-dimensional array is a simple list of elements, while a two-dimensional array is like a table with rows and columns. Multi-dimensional arrays can have more than two dimensions, forming a higher-dimensional structure.

Arrays have several key properties:

1. Size: An array has a fixed size, which is determined when it is created. The size represents the total number of elements that can be stored in the array.

2. Element Type: All elements in an array must be of the same type, such as integers, characters, or objects. This allows for efficient memory allocation and retrieval.

3. Random Access: Elements in an array can be accessed directly using their indices. This enables fast and constant-time access to any element in the array.

4. Contiguity: Array elements are stored in contiguous memory locations. This property ensures that accessing neighboring elements in the array is efficient.

Overall, arrays are a fundamental and powerful tool in programming that provide efficient storage and retrieval of homogeneous data. They are widely used and supported in most programming languages and play a vital role in many algorithms and applications.



#  Linked Lists

A linked list is a linear data structure used in computer science and programming. Unlike arrays, which store elements in contiguous memory locations, linked lists store elements in separate nodes that are linked together through pointers or references.

Each node in a linked list contains two components: the data and a reference to the next node in the sequence. The first node in the list is called the head, and the last node typically points to null, indicating the end of the list.

Here are some key characteristics and features of linked lists:

1. Dynamic Size: Linked lists have a dynamic size, meaning they can grow or shrink as elements are added or removed. This flexibility is in contrast to arrays, which have a fixed size.

2. Insertion and Deletion: Inserting or deleting an element in a linked list is typically more efficient than in an array because it involves updating pointers rather than shifting elements in memory. However, accessing a specific element in a linked list takes linear time since the list must be traversed sequentially.

3. No Contiguity: Unlike arrays, linked lists do not require contiguous memory allocation. Each node can be located anywhere in memory, allowing for efficient memory utilization.

4. Singly Linked List vs. Doubly Linked List: In a singly linked list, each node contains a reference to the next node. In a doubly linked list, each node has references to both the next and previous nodes. The use of doubly linked lists allows for efficient traversal in both directions but requires additional memory for storing the backward references.

It's important to note that linked lists have some disadvantages compared to arrays. They require additional memory for storing the references, and accessing a specific element requires traversing the list sequentially. Additionally, linked lists do not provide constant-time random access like arrays.

Overall, linked lists provide a flexible and efficient way to store and manipulate collections of data. They are widely used in programming and serve as a foundation for more complex data structures and algorithms.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def printList(self):
        temp = self.head
        while temp:
            print(temp.data)
            temp = temp.next

    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    def insertAfter(self, prev_node, new_data):
        if prev_node is None:
            print("The node does not exist")
            return
        new_data = Node(new_data)
        new_data.next = prev_node.next
        prev_node.next = new_data

    def append(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def delete_node(self, key):
        temp = self.head
        if temp and temp.data == key:
            self.head = temp.next
            temp = None
            return

        while temp:
            if temp.data == key:
                break
            prev = temp
            temp = temp.next

        if temp == None:
            return
        prev.next = temp.next
        temp = None
```

# Selection Sort

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
        
        
a=["c","b","x"]
selection_sort(a)
```

1.The selection_sort function is defined, which takes an array arr as an input.

2.The variable n is assigned the length of the array using the len() function. This will be used to determine the number of elements in the array.

3.The outer loop iterates over the array from the first element to the last element. It uses the variable i as the index to keep track of the sorted part of the array.

4.Inside the outer loop, a variable min_index is initialized to i. This variable will keep track of the index of the minimum element in the unsorted part of the array.

5.The inner loop starts from i+1 and iterates to the end of the array. It searches for the minimum element by comparing each element with the current minimum element (arr[min_index]). If a smaller element is found, the min_index is updated to the index of that element.

6.After the inner loop completes, the minimum element in the unsorted part of the array is found. Now, we swap the minimum element with the first element in the unsorted part of the array. This brings the minimum element to its correct sorted position.

7.This swapping is done using a Pythonic way of swapping values: arr[i], arr[min_index] = arr[min_index], arr[i].

8.The outer loop continues, and the process repeats until the entire array is sorted.

9.Finally, the sorted array is returned.



This code demonstrates the selection sort algorithm by sorting an array of integers in ascending order. The smallest element is repeatedly selected from the unsorted part of the array and placed in its correct position in the sorted part of the array.
